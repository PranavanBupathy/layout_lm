```python
def item_generator(json_input, lookup_key):
    if isinstance(json_input, dict):
        for k, v in json_input.items():
            if k == lookup_key:
                yield v
            else:
                yield from item_generator(v, lookup_key)
    elif isinstance(json_input, list):
        for item in json_input:
            yield from item_generator(item, lookup_key)
```

```python
import pandas as pd
from datasets import Dataset, DatasetDict, Features, Array2D, Value, ClassLabel

# Load your data into a pandas DataFrame
df = pd.read_csv('your_data.csv')

# Define the features for your dataset
features = Features({
    'id': Value('int64'),
    'words': Array2D(dtype='string', shape=(None,)),
    'bboxes': Array2D(dtype='float32', shape=(None, 4)),
    'ner_tags': Array2D(dtype='int64', shape=(None,)),
    'image_path': Value('string'),
})

# Split your data into train and test sets
train_data = df[:100]
test_data = df[100:150]

# Create Hugging Face Datasets from your data
train_dataset = Dataset.from_pandas(train_data, features=features)
test_dataset = Dataset.from_pandas(test_data, features=features)

# Create a DatasetDict containing your train and test datasets
dataset_dict = DatasetDict({
    'train': train_dataset,
    'test': test_dataset,
})
```

```python
import json
from pycocotools.coco import COCO
from datasets import Dataset, DatasetDict

# Load the COCO-formatted dataset
coco = COCO('/home/ubuntu/Layoutlm_Finetune_Inference/Data/result.json')

# Convert the COCO dataset to the desired format
train_data = Dataset.from_dict({
    'id': [str(img['id']) for img in coco.imgs.values()],
    'words': [ann['words'] for ann in coco.anns.values() if coco.imgs[ann['image_id']]['split'] == 'train'],
    'bboxes': [ann['bbox'] for ann in coco.anns.values()],
    'ner_tags': [ann['ner_tags'] for ann in coco.anns.values() if coco.imgs[ann['image_id']]['split'] == 'train'],
    'image_path': [coco.loadImgs(img['id'])[0]['file_name'] for img in coco.imgs.values() if img['split'] == 'train']
})

test_data = Dataset.from_dict({
    'id': [str(img['id']) for img in coco.imgs.values() if img['split'] == 'test'],
    'words': [ann['words'] for ann in coco.anns.values() if coco.imgs[ann['image_id']]['split'] == 'test'],
    'bboxes': [ann['bbox'] for ann in coco.anns.values() if coco.imgs[ann['image_id']]['split'] == 'test'],
    'ner_tags': [ann['ner_tags'] for ann in coco.anns.values() if coco.imgs[ann['image_id']]['split'] == 'test'],
    'image_path': [coco.loadImgs(img['id'])[0]['file_name'] for img in coco.imgs.values() if img['split'] == 'test']
})

# Combine the training and testing data into a DatasetDict
data_dict = DatasetDict({'train': train_data, 'test': test_data})
```
